```python
# Goal => Predicting Salary of a new employee based on years of experience. 
```

```python
# Import Libraries 

# Numpy ?? =>  open source math numerical python library, contains multi-dimensional arrays and matrix data structures.
# Used to perform mathematical operations on arrays. 

# Pandas ?? => open source python library for data manipulation and analysis. 

# Matplotlib ?? => data visualisation library for python. 

# Seaborn ?? => data visualisation library for python. 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
# Getting the Dataset which is in csv form. 
# csv ?? => comma seperated values is a simple format used to store tabular data ie spreadsheets. 
salaryDataset = pd.read_csv('Salary_Data.csv')
```

```python
# take a peak into the dataset. 
# Observations => looking at the dataset, can see years of experience increases, salary also increases. 
salaryDataset.head(10) 
```

```python
# Some Statistical description of the dataset, looking at the mean salary, minimum salary and maximum salary. 
salaryDataset.describe() 
```

```python
# Exploratory Data Analysis => Visualise this simple dataset. 
# Observe that there is correlation betweeb salary and experience. 
sns.pairplot(salaryDataset)
```

```python
# building a Simple Linear Reegression Model 
# Splitting the dataset into training and testing
X = salaryDataset.iloc[:, :-1].values 
y = salaryDataset.iloc[:, -1].values 
```

```python
# Print Independent variable X and observe 
print(X)
```

```python
# Print independent variable y and observe 
print(y) 
```

```python
# Split the dataset to training and test set using scikit learn. 
# Scikit Learn ?? => open sourced machine learning library that supports supervised and unsupervised learning.
# train_test_split => splitting the dataset into so model can train on some and test on some. 
from sklearn.model_selection import train_test_split
```

```python
# Independent variables split to training and test set, dependent variables split to training and test set. 
# test_size?? => part of the dataset set aside for testing the model. 30% set for testing. 
# random_state?? => controls the shuffling applied to the data before the split.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
# Training simple linear regression model 
from sklearn.linear_model import LinearRegression # import first 
simpleregressor = LinearRegression() # instance of linear regression model 
```

```python
# fit the object into training data 
simpleregressor.fit(X_train, y_train)
```

```python
# Predict the test results 
salary_prediction = simpleregressor.predict(X_test) # predict salary based on years of experience 
print(salary_prediction)
```

```python
# Visualise the Training set results using matplotlib 
plt.scatter(X_train, y_train, color="red") 
plt.plot(X_train, simpleregressor.predict(X_train), color="blue") # plot training set(years of experience) and predicted salary using line color blue
plt.title('Salary vs Experience(Training set')
plt.xlabel('Years of Experience')
plt.show() 
```

```python
# Visualising the Test set results 
plt.scatter(X_test, y_test, color="red") # scatter plot of testing set(X) and test set (y(output)), 
plt.plot(X_train, simpleregressor.predict(X_train), color="blue") # plot training set(X) and predicted salary using color blue
plt.title('Salary vs Experience(Test set)') 
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show() 
```

```python
# Regression Evaluation Metrics 
from sklearn import metrics 
```

```python
print('Mean Squared Error: ', metrics.mean_squared_error(y_test, salary_prediction))
print('Mean Absolute Error: ', metrics.mean_absolute_error(y_test, salary_prediction))
print('Root Mean Squared Error: ', np.sqrt(metrics.mean_squared_error(y_test, salary_prediction)))
```

```python
# Interpretations 
# Mean Squared Error => tells you how close regression line is to a set of points, takes distances from pts to the regression line and squares them. 
# The lower the Mean Squared, the better the forecast. 

# Mean Absolute Error => measures the average magnitude of errors in a prediction without considering direction.
# similar to MSE. MAE takes the sum of absolute value of error. MAE is a more direct representation of error terms. 

# RMSE => square root of MSE, MSE is square of error and square root brings it back to same level of prediction 
# error and makes it easier to interpret.
```
